# Método polinomial

Sejam $a(s)$ e $b(s)$ dois polinomios coprimos, com graus $n_a$ e $n_b$, onde $n_a > n_b$. Para qualquer polinômio arbitrário $a^*(s)$ com grau $n^*_a \geq n_a$ vale:

$$a(s)l(s)+b(s)p(s)=a^*(s)$$

As soluções $l(s)$ e $p(s)$ são únicas e 

$$
\left\{\begin{array}
& n_l \leq \text{max}{(n^*_a-n_a, n_b-1)}\\
n_p < n_a
\end{array}\right.
$$

### Projeto de controladores

$$G_c(s)=\frac{p(s)}{l(s)}, G_p(s)=\frac{b(s)}{a(s)}$$

$$FTMF: \frac{Y(s)}{R(s)}=\frac{p(s)b(s)}{a(s)l(s)+b(s)p(s)}=\frac{p(s)b(s)}{a^*(s)}$$

$a^*(s)$ deve introduzir polos que levem o sistema ao desempenho desejado. Se houver necessidade de melhorar erro de regime, polos devem ser introduzidos em $G_p(s)$ original, e, ao final, remanejados de volta para o controlador.

1. Método I: o controlador é obtido automaticamente
  
  a) Escolher $a^*(s)\Rightarrow n_a^* \geq n_a$
  
  b) Encontrar $n_l \leq \text{max}{(n_a^*-n_a,n_b-1)}$
  
  c) $n_p < n_a$
  
2. Método II: pressupor o tipo de controlador

  a) Escolher controlador
  
  b) De posse de $n_l, n_p$, escolher $n_a^* = \text{max}(n_a+n_l,n_b+n_p)$
  
  c) Escolher $a^*(s)$ com base no desempenho desejado

## Projeto de controladores usando resposta em frequencia

### Definições preliminares

- Frequência de corte: aquela para a qual a atenuação é maior igual a 3 dB

- Banda passante $$0 \leq \omega \leq \frac{1}{\tau}$$

- $\text{Banda larga }\Rightarrow \text{resposta rápida}\\\text{Banda estreita }\Rightarrow \text{resposta lenta}$



### Funções transferência dos controladores

1. PD

$$G_c(s)=K_c(1+\tau_d s)$$

  - Melhora a marge de fase

2. PI

$$G_c=\frac{(K_c/\tau_i)\left(1+\tau_i s\right)}{s}$$

  - Piora a marge de fase
  
3. PID

$$\tau_i = 4\tau_d\ (\text{zeros reais e iguais})$$

$$G_c(s) = \frac{(K_c/4\tau_d)(2\tau_d s+1)^2}{s} $$

  - Melhora margem de fase
  - Melhora erro de regime

### Exemplos (Projetos usando controladores por métodos frequenciais)

#### 1. PD com especificação de erro fixada

$$\text{Especificações}\left\{\begin{array}\ MF \geq 70 \deg \\ MG \geq 6\text{ dB}\\\text{e.r} = 0.05, r(t)=t\\\omega_B\geq 6.3\text{ rad/s} \end{array}\right.$$

$$G_p(s)=\frac{40}{s(s+2)}$$

Checando desempenho:

Fixando-se $\text{e.r} = 0.05, r(t)=t, $

$$\text{e.r}=\frac{1}{K_v}=\frac{1}{\lim_{s\rightarrow 0}s[G_p(s)G_c(s)]}=0.05$$

$$\therefore K_c = 1$$

Portanto, $G_c(s)=1+\tau_d s.$

Se eu pudesse escolher o $K_c,$ garantiria $ \phi_c(6.3)=52\deg$ (déficit mantendo frequencia de corte). Como o $K_c$ é fixo, preciso deslocar a curva do PD para esquerda, o que ocasiona aumento de $fcg$ (PD dá ganho positivo) e diminuição de $M_F$. Por causa disso, o PD tem que compensar com um ângulo maior do que teria na situação em que $fcg$ se mantesse constante, e isso é expresso em

$$\phi_c'=52\deg+\varepsilon$$

- 1ª tentativa: $ \varepsilon = 5\deg \therefore \phi_c' = 57 \deg $

$$\phi_C' = \arctan(\tau_d \omega)\Rightarrow \tau_d \omega = \tan{\phi_C'} \Rightarrow \tau_d = \frac{\tan{\phi_C'}}{\omega}=\frac{1.54}{\omega}$$

Ganho do PD na frequência $\frac{1.54}{\omega}:$

$$20=\frac{A}{\log{\frac{1.54}{\tau_d}}-\log{\frac{1}{\tau_d}}}\Rightarrow A = 3.8 \text{ dB}.$$

Encontrando $\omega$ para o qual  $|G_p(s)|=-3.8\text{ dB},$ encontra-se  a nova $fcg$.

$$ G_p(j\omega) = \frac{40}{j\omega(j\omega+2)}=\frac{40}{-\omega^2+j2\omega} $$

$$|G_p(j6.3)|=0.961 = -0.346 \text{ dB}$$

$$-40 = \frac{-0.346-(-3.8)}{\log{6.3}-\log{\omega_B}}\Rightarrow \omega_B = 7.69 \text{ rad/s} \therefore G_c(s) = 1+0.2s$$

$$G(s) = G_p(s)G_c(s)=\frac{40(1+0.2s)}{s(s+2)}\Leftrightarrow G(j\omega)=\frac{40(1+j0.2\omega)}{j\omega(j\omega+2)}$$

$$\therefore \angle{G(j7.69)}=-108\deg \Rightarrow M_F = 72 \deg, $$ o que satisfaz às especificações de desempenho.

#### 2. PD sem especificação de erro fixada

Mesmas especificações do exemplo anterior, exceto que $\text{e.r}=0,r(t)=u(t)$.

Como a $fcg$ se mantém (porque posso ajustar o $K_c$):

$$G_c(j\omega)=K_c(1+j\omega\tau_d ) \Rightarrow \angle{G_c(s)}=\arctan{\omega \tau_d}=d=52\deg\\ \therefore \tan{d}=\omega\tau_d \Leftrightarrow \tau_d = \frac{\tan{52}}{6.3}\approx 0.2$$

Agora, usando o $K_c$, ajusto o PD para não influir na $fcg$ (ganho do PD deve ser zero na $fcg$ escolhida):

$$20=\frac{0-K_c(\text{dB})}{\log{\frac{1}{6.3}}-\log{\frac{1}{\tau_d}}} \Rightarrow K_c(\text{dB}) = 20\left(\log{\frac{1}{\tau_d}}-\log{\frac{1}{6.3}}\right) \Rightarrow K_c = 0.79 $$

Assim, $G_c(s)=0.79(1+0.2s)$

#### 3. PI

$$G_p(s) = \frac{0.5}{s(s+1)(0.5s+1)}\left\{\begin{array}\ MF\geq 40\deg\\MG\geq 6\text{ dB}\\\omega_B=0.5\text{ rad/s}\\\text{e.r}=0,r(t)=t\end{array}\right.$$

Checando especificações:

a) $MF$

$$\angle{G_p(j\omega)\big{|}_{\omega=\omega_B}}=-130\deg \Rightarrow MF=180-130=50\deg$$

b) $MG$

$$fcf = \sqrt{2}\text{ rad/s}\therefore |G_p(j\omega)|\big{|}_{\omega=\text{fcf}}=-12\text{ dB} \Rightarrow MG=12\text{ dB} $$

c) $\text{e.r}$

$$\text{e.r}=\frac{1}{K_v}=\frac{1}{\lim_{s\rightarrow 0}sG_p(s)}=2$$

Concl.: problema no erro de regime me leva a concluir que devo usar um controlador PI para deixar o sistema dentro das especificações que eu quero. Logo,

$$G_c(s)=\frac{K_c/\tau_i(1+\tau_i s)}{s}$$

Como o controlador PI piora a margem de fase, a ideia é colocar o $1/\tau_i$ bem à esqueda do $\omega_B$ desejado. 

---

*Regra empírica:* $$\frac{1}{\tau_i}=\frac{1}{a}\cdot\text{fcg},\ 10\leq a\leq20$$

---

1ª tentativa: $a = 10 \Rightarrow 1/\tau_i=0.1\cdot 0.5=0.05\therefore \tau_i=20$

$$ G_c(s) = \frac{0.05K_c(20s+1)}{s} $$

  - O problema do erro de regime foi resolvido. Resta checar $MG, MF.$

a) $MF$

$$ \angle{G_c(s)}\big{|}_{\omega=\omega_B}=-90+\arctan{(20\cdot 0.5)}=-5.71 \deg  \therefore MF = 180+(-130-5.71) = 44.3 \deg $$

b) $MG$
 
Posso manter a margem de ganho encontrada anteriormente ($12\text{ dB}$) ajustando o $K_c$ de modo que $|G_c(j\omega)|\big{|}_{\omega=fcf}$ seja 1. Sabe-se que o $fcf$ sem o PI era $\sqrt{2}$. Como o PI influiria com ângulo desprezível em $\omega=\sqrt{2}$, $fcf$ se mantém. Logo,

$$ G_c(j\omega) = \frac{0.05K_c(j20\omega+1)}{j\omega}\Leftrightarrow |G_c(j\omega)|=\frac{0.05K_c\sqrt{400\omega^2+1}}{\omega}$$

$$K_c = \frac{\omega}{0.05\sqrt{400\omega^2+1}} = 1  $$

Concl.: margem de fase e de ganho continuam dentro das especificações de projeto. 

Adicional: qual $K_c$ mínimo pra $MG$ ainda estar de acordo com as especificações de projeto?

$|G_c(s)|=\text{6 dB}$ para $MG=-6\text{ dB} $

Assim, 

$$K_c = \frac{2\omega}{0.05\sqrt{400\omega^2+1}} \therefore K_c = 2 $$





#### 4. PID

$$G_p(s) = \frac{1}{s(s+1)(0.5s+1)}\left\{\begin{array}\ MF\geq 45\deg\\MG\geq 6\text{ dB}\\\omega_B=\sqrt{2}\text{ rad/s}\\\text{e.r}\leq0,r(t)=t\end{array}\right.$$

Checando especificações:

a) $MF$

$$\begin{align}
1& = |G_p(s)|\\
&= \frac{1}{|j\omega||j\omega+1||j0.5\omega +1|}\\
&= \frac{1}{\omega\sqrt{0.25\omega^4+1.25\omega^2+1}}\\
&\therefore 0.25\omega^6+1.25\omega^4+\omega^2-1=0\\ 
&\Rightarrow \omega_B \approx 0.72 \text{ rad/s} = fcg
\end{align}$$

$$\angle{G_p(j\omega)}\Big{|}_{\omega=fcg}=-90-\tan^{-}{(\omega)}-\tan^{-}{(0.5\omega)}=-146^{\circ} \therefore MF = 34^{\circ}$$

b) $MG$

$$G_p(j\omega)=\frac{1}{-1.5+j(\omega-0.5\omega^3)}=\frac{-1.5-j(\omega-0.5\omega^3)}{1.5^2+(\omega-0.5\omega^3)^2}$$

$$\omega-0.5\omega^3=0\Rightarrow \omega=\sqrt{2}\text{ rad/s}=fcf$$

$$|G_p(j\omega)|\Big{|}_{\omega=fcf}=-9.54\text{ dB}\Leftrightarrow MG=9.54\text{ dB}$$

c) $\text{e.r}$
 
 $$\text{e.r}=\frac{1}{K_v}=\frac{1}{\lim_{s\rightarrow 0}sG_p(s)}=1$$
 
 Precisa-se corrigir:
  
  - erro de regime
  - margem de fase
  - $\omega_B$
  
Solução: PID!

$$G_c(s)=\frac{K_c/4\tau_d(2\tau_d s+1)^2}{4\tau_d s}$$

Correções:

a) $M_F$

$$\angle{G_p(j\omega)}\Big{|}_{\omega=fcg}=-180^{\circ} \therefore MF=0$$

Concl.: o controlador fornece sozinho os $45^{\circ}$. Assim,

$$\angle{G_c(j\omega)}\Big{|}_{\omega=\sqrt{2}}=-90+2\arctan{(2\tau_d \omega)}=45$$

$$\therefore \tau_d\approx 0.88$$

b) $\omega_B$

O controlador deve contribuir, em $\omega=\sqrt{2}\text{ rad/s} $ com um ganho oposto ao de $G_p(s)$

$$|G_p(j\omega)|\Big{|}_{\omega=\sqrt{2}}=-9.54\text{ dB}\Rightarrow  |G_c(j\omega)|\Big{|}_{\omega=\sqrt{2}}=9.54\text{ dB}=3.0$$

$$\therefore \frac{K_c(2\tau_d s+1)^2}{4\tau_d s}=3.0 \Rightarrow K_c \approx 2.08$$

Finalmente,

$$G_c(s)=\frac{0.59(1.76s+1)^2}{s}$$
